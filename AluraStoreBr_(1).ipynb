```python
def grafico_produtos_por_loja(loja, nome_loja):
    produtos = [produto for produto, _ in loja]
    vendas = [vendas for _, vendas in loja]

    # Organizando as listas por vendas (menor para maior)
    produtos_ordenados = [produto for _, produto in sorted(zip(vendas, produtos))]
    vendas_ordenadas = sorted(vendas)

    # Criando o gráfico de barras para a loja
    plt.figure(figsize=(10, 6))
    plt.barh(produtos_ordenados, vendas_ordenadas, color='skyblue')

    # Adicionando título e rótulos
    plt.title(f'Produtos Mais e Menos Vendidos - {nome_loja}')
    plt.xlabel('Quantidade de Vendas')
    plt.ylabel('Produtos')

    # Exibindo os valores das vendas no gráfico
    for i in range(len(vendas_ordenadas)):
        plt.text(vendas_ordenadas[i] + 0.1, i, f'{vendas_ordenadas[i]}', va='center', fontsize=10)

    # Exibindo o gráfico
    plt.tight_layout()
    plt.show()

# Dados de produtos e vendas por loja
lojaa = [("Assistente virtual", 1), ("Mesa de janta", 1), ("Jogo de tabuleiro", 1), ("Micro-ondas", 1), ("Cadeira de escritório", 1)]
lojaa2 = [("Cadeira de escritório", 2),("Cama King", 2), ("Pandeiro", 1)]
lojaa3 = [("Secadora de roupas", 1), ("Cômoda", 1), ("Violão", 1), ("Bola de vôlei", 1), ("Guarda roupas", 1)]
lojaa4 = [("Lava louças", 1), ("Cama box", 1), ("Bola de basquete", 1), ("Modelagem preditiva", 1), ("Fone de ouvido", 1)]

# Gerar o gráfico para cada loja separadamente
grafico_produtos_por_loja(lojaa, 'Loja 1')
grafico_produtos_por_loja(lojaa2, 'Loja 2')
grafico_produtos_por_loja(lojaa3, 'Loja 3')
grafico_produtos_por_loja(lojaa4, 'Loja 4')
```
